```python
%matplotlib inline
import matplotlib.pyplot as plt
fig=plt.figure(dpi= 120, facecolor='w', edgecolor='k')
#### read in the txt file
import numpy as np
raw_data = np.loadtxt('../data/ur5_translated_pose_data.txt', delimiter=',')

trunk = raw_data[40:170,:]

low_index = np.argmin(trunk[:,6],axis=0)
shift = np.array(trunk[low_index:(low_index + 1),:])
shift[0,0] = 0
shift[0,4:] = 0
translated_data = trunk - np.repeat(shift, trunk.shape[0], axis=0)
plt.plot(translated_data[:,1:])
low_index
np.savetxt('../data/calibrated_pickup.txt', translated_data, fmt="%0.10f", delimiter=', ')
```
